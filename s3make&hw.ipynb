```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('download.png')
```

```python
b , g , r = cv2.split(image)
```

```python
eq_b = cv2.equalizeHist(b)
eq_g = cv2.equalizeHist(g)
eq_r = cv2.equalizeHist(r)
```

```python
eq_image = cv2.merge([eq_b,eq_g,eq_r])
```

```python
plt.subplot(121)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.xticks([]),plt.yticks([])

plt.subplot(122)
plt.imshow(cv2.cvtColor(eq_image,cv2.COLOR_BGR2RGB))
plt.title("Equalized Image")
plt.xticks([]),plt.yticks([])
```

```python
noise_image = cv2.imread("1631302689903.jpg")
```

```python
filtered_image = cv2.GaussianBlur(noise_image,(5,5),0)
```

```python
plt.subplot(121)
plt.imshow(noise_image,cmap="gray")
plt.title("Original Image")
plt.xticks([]),plt.yticks([])

plt.subplot(122)
plt.imshow(filtered_image,cmap="gray")
plt.title("Filtered Image")
plt.xticks([]),plt.yticks([])
```

```python
median_image = cv2.medianBlur(noise_image,5)
```

```python
plt.subplot(121)
plt.imshow(noise_image,cmap="gray")
plt.title("Original Image")
plt.xticks([]),plt.yticks([])

plt.subplot(122)
plt.imshow(median_image,cmap="gray")
plt.title("Filtered Image")
plt.xticks([]),plt.yticks([])
```
